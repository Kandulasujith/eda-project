```python
data_lst=[]
for i in range(42,47):
    url2=f"https://www.amazon.in/s?k=helmet+for+bike&i=automotive&rh=n%3A4772060031&dc&page={i}&crid=2VXBO7EFL5APE&qid=1741597992&rnid=91049095031&sprefix=%2Caps%2C1563&xpid=Fii8rnlyDwkfN&ref=sr_pg_{i}"
    webpage = requests.get(url2,headers = HEADERS)
    soup = BeautifulSoup(webpage.content,'html.parser')
    links = soup.find_all('a',attrs = {'class':"a-link-normal s-line-clamp-4 s-link-style a-text-normal"})
    link_list =[]
    for link in links:
        link_list.append(link.get('href'))

    for link in link_list:
    # Introduce a delay to prevent rate-limiting
        time.sleep(random.uniform(2, 5))  # Random delay between 2 to 5 seconds

        webpages = requests.get("https://www.amazon.in/" + link, headers=HEADERS)
        soups = BeautifulSoup(webpages.content, "html.parser")

        try:
            data = soups.find("table", attrs={"class": "a-normal a-spacing-micro"}).text
            price = soups.find("span", attrs={"class": "a-price-whole"}).text
            rating_count = soups.find("span", attrs={"id": "acrCustomerReviewText"}).text
            ratings_counts=re.findall(r"\d+,\d+",rating_count)
            rating = soups.find("span", attrs={"class":"a-icon-alt"}).text
            ratings=re.findall(r"\d.\d+",rating)
           # prating=soups.find("img",attrs = {"class":"brand-snapshot-item-image"}).text
            brand_info = soups.find_all("div",class_="a-section a-spacing-base brand-snapshot-flex-row")
            if brand_info:
                p = brand_info[0].find("p").text.strip()
                prating=re.findall(r"(\d+\%)",p)
            else:
                prating = "N/A"
            
            

        
        except AttributeError:
          
            data,price, rating_count, rating,prating = None, None, None, None,None

        data_columns = ["Brand", "Colour", "Item Weight", "Material"]
        pattern = r"Colour\s+([\w\s()-]+?)\s+Brand\s+([\w\s()-]+?)\s+Vehicle Service Type.*?Item Weight\s+([\w\s()-]+?)\s+Material\s+([\w\s()-]+)"

   

        match = re.search(pattern, data) if data else None

        values_list = [match.group(i).strip() if match else "N/A" for i in range(1, 5)]
        values_list.append(rating_count if rating_count else "N/A")
        if price:
            if ',' in price:
                p1=re.findall(r"\d+,\d+",price)
                values_list.append(p1)
    
            else:
                p1=re.findall(r"\d+",price)
                values_list.append(p1)
        else:
            values_list.append('N/A')

        values_list.append(ratings[0] if rating else "N/A")
        values_list.append(prating[0] if prating else "N/A")
        
        data_lst.append(values_list)
```
